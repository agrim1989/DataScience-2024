# Importing Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

### Importing Dataset

```
df = pd.read_csv('covid.csv', encoding='ISO-8859-1')
df.dropna()
df.head()
```

### adding percentage for Recovered and Death

```
df['Recovered_Percentage'] = round((df['Recovered']/df['Cases'])*100, 2)
df['Death_Percentage'] = round((df['Deaths']/df['Cases'])*100, 2)
df.head()
```

### finding higest recovered person by counrties

```
df.sort_values(by=['Recovered_Percentage'], ascending=False).head()
```

### Make a graph for top 5 Recovered Percentage

```
# prompt: make a graph for top 5 Countries Name with Recovered Percentage

df.sort_values(by=['Recovered_Percentage'], ascending=False).head().plot(kind='bar', x='Country Name', y='Recovered_Percentage')
plt.title('Top 5 Countries with Highest Recovered Percentage')
plt.xlabel('Country')
plt.ylabel('Recovered Percentage (%)')
plt.xticks(rotation=90)
plt.show()
```

### Sorted data by Death in country

```
df.sort_values(by=['Deaths'], ascending=False).head()
```

### Graph top 5 Counrey Name with Deaths

```
df.sort_values(by=['Deaths'], ascending=False).head().plot(kind='bar', x='Country Name', y='Deaths')
plt.title('Top 5 Countries with Highest Deaths')
plt.xlabel('Country')
plt.ylabel('Deaths')
plt.xticks(rotation=90)
plt.show()
```

### Sorting by Death Percentage

```
df.sort_values(by=['Death_Percentage'], ascending=False).head()
```

```
df.sort_values(by=['Death_Percentage'], ascending=False).head().plot(kind='bar', x='Country Name', y='Death_Percentage')
plt.title('Top 5 Countries with Highest Deaths')
plt.xlabel('Country')
plt.ylabel('Deaths')
plt.xticks(rotation=90)
plt.show()
```
